```python
import os

dataset_path = os.path.join('data')
words_path = os.path.join(dataset_path, 'words')
clues_path = os.path.join(dataset_path, 'clues')
puzzles_path = os.path.join(dataset_path, 'puzzles')
```

```python
f = open(words_path)
lines = f.readlines()

words = {}
n = 0
max_n = 0
for line in lines:
    if n >= max_n and max_n > 0:
        break
    
    word = line.strip()
    length = len(word)
    
    if not length in words:
        words[length] = []
    
    words[length].append({ 'word': word, 'line': n})
    
    n += 1

print(len(words))
```

```python
import random
import numpy as np

debug_mode = False

def create_puzzle(width, height, seed = None, illegal_down_columns = 0, illegal_down_words = 0, r = 0):
    r += 1
    
    if r >= 100:
        return
    
    if illegal_down_columns >= width:
        if illegal_down_words >= len(words[height]):
            return
        return create_puzzle(width, height, seed, 0, illegal_down_words + 1, r)
    
    if illegal_down_words >= len(words[height]):
        return
    
    if seed == None:
        seed = random.randrange(100)
    
    np.random.seed(seed)
    random.seed(seed)
    
    down_index = np.random.permutation(len(words[height]))[illegal_down_words]
    down = words[height][down_index]
    
    down_column = np.random.permutation(width)[illegal_down_columns]
    across = {}
    
    down_chars = [char for char in down['word']]
    word_indexes = []
    
    if width == height:
        word_indexes.append(down_index)
    
    for w in np.random.permutation(len(words[width])):
        if width == height and w == down_index:
            continue
        
        word = words[width][w]
        word_str = word['word']
        if word_str[down_column] in down_chars:
            word_indexes.append(w)
            idx = down_chars.index(word_str[down_column])
            down_chars[idx] = ''
            
            across[str(idx)] = word
            
    if len(across.keys()) == height:
        if not debug_mode:
            for w in sorted(word_indexes, reverse=True):
                del words[width][w]
            if height != width:
                del words[height][down_index]
        return {
            'across': dict(sorted(across.items())),
            'down': down,
            'down_column': int(down_column),
            'height': height,
            'index': height * width,
            'width': width
        }
    
    illegal_down_columns += 1
    return create_puzzle(width, height, seed, illegal_down_columns, illegal_down_words, r)
```

```python
import json

def save_puzzle(puzzle):
    f = open(puzzles_path, 'a')
    
    if os.stat(puzzles_path).st_size == 0:
        f.write('[')
    else:
        f.write(',')
    
    f.write(json.dumps(puzzle, indent = 4))
```

```python
if os.path.exists(puzzles_path):
    os.remove(puzzles_path)

puzzle_count = 0
for height in sorted(words.keys()):
    if height <= 0:
        continue
        
    n = 0
    for width in sorted(words.keys()):
        if width <= 0:
            continue
        
        print(height, 'x', width)
        
        puzzle = ''
        n = 0
        while puzzle != None:
            n += 1
            if n % 10 == 0:
                print('n', n)
                
            puzzle = create_puzzle(width, height)

            if puzzle != None:
                puzzle_count += 1
                save_puzzle(puzzle)
            else:
                break

f = open(puzzles_path, 'a')
f.write(']')
f.close()

print(puzzle_count)
```
